## Fix image file names in json label Label Studio export file
#### Label Studio exports have cloud bucket/blob name as prefix to file
#### Must remove this to use for training in local 

```python
import json

# with open('/home/lstm/github/PaddleDetection/dataset/_KRCASE_20240516/labels.json', 'r') as file:
with open('/home/lstm/github/PaddleDetection/dataset/_KRCASE_20240516/result.json', 'r') as file:
    json_data = json.load(file)

json_data.keys()
```

```python
# review categories
json_data['categories']
```

```python
# convert id values from str to int
for idx, el in enumerate(json_data['categories']):
    if isinstance(el['id'], str):
        el['id'] = int(el['id'])

# review conversion
json_data['categories']
```

```python
# inspect file name
json_data['images']
```

```python
# fix file name
org_str = 'azure-blob://cases'
new_str = 'images'  # this is folder name in the 'train' directory in the root of the dataset

fixed_names = [{**el, 'file_name': el['file_name'].replace(org_str, new_str)} for el in json_data['images']]

fixed_names
# json_data['images']
```

```python
# reaplace json_data['images'] with fixed_names
json_data['images'] = fixed_names
```

```python
# review annotations
json_data['annotations']
```

```python
# convert category_id values from str to int
for idx, el in enumerate(json_data['annotations']):
    if isinstance(el['category_id'], str):
        el['category_id'] = int(el['category_id'])

# review conversion
json_data['annotations']
```

```python
# save json_data to file
with open('/home/lstm/github/PaddleDetection/dataset/_KRCASE_20240516/labels_fixed.json', 'w') as file:
    json.dump(json_data, file, indent=4)

# verify fix
with open('/home/lstm/github/PaddleDetection/dataset/_KRCASE_20240516/labels_fixed.json', 'r') as file:
    verify_data = json.load(file)
verify_data['images']
```
