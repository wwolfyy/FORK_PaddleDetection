# Get a list of documents that have a particular element

```python
import os, shutil

source_name_folder = './testresults/20240528/test_set_footnote_20240528_threshold_0_1'
source_file_folder = './testdata/test_set_footnote'
target_folder = '/home/lstm/github/unstructured-inference/DATA/sample_data/판례/footnote'
```

```python
# get list of files (that have element in question)
filelist = os.listdir(source_name_folder)

# copy file from source file folder to target folder
for filename in filelist:
    if filename.endswith('.jpg'):
        shutil.copy2(f'{source_file_folder}/{filename}', f'{target_folder}/{filename}')
```
